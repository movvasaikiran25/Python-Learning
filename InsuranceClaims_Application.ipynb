```python
class Claims:
    def __init__(self, type_code,claim_totalbalance):
        self.__type_code = type_code  # private attribute
        self.__claim_totalbalance = claim_totalbalance
        type_code = input("Enter the claim type code: ")
        if type_code not in ["AUTO", "HOME", "HEALTH"]  :
            raise ValueError("Invalid claim type code.")
        else:
          print (f"Claim Type Code: {type_code}")
    def reserve_amount(self, amount):
        if amount <=100 :
            self.__claim_totalbalance += amount
            return self.__claim_totalbalance
        else:
            return "Reserve amount exceeds limit."
    def payment_amount(self, amount):
            if 0 < amount >= self.__claim_totalbalance:
                self.__claim_totalbalance -= amount
                return self.__claim_totalbalance
            else:
                return "Payment amount exceeds total balance."
    def get_total_balance(self):
            return self.__claim_totalbalance

claim = Claims("AUTO", 5000)

try:
    claim.__claim_totalbalance +=1000
except AttributeError :
    print("Cannot access private attribute directly.")

print ("The claim total balance", claim.get_total_balance())
claim.reserve_amount(80)
print ("The claim total balance after reserving amount:", claim.get_total_balance())
claim.payment_amount (200)
print ("The claim total balance after payment amount:", claim.get_total_balance())
```
